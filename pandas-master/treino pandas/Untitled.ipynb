GRAFICO BASICO

```python
x = [1,8,22,6,11]
y = [1,4,10,12,1]

#mudando tamanho do grafico
plt.figure(figsize=(5,3), dpi=300)

#Versão longa
#plt.plot(x,y, label='ih ala uma legenda', color='orange', linewidth = 2,marker='.')

#versão curta
plt.plot(x,y, 'b^--', label='2x')

#Outra Linha
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:5],x2[:5]**2,'r',label='x²')
plt.plot(x2[4:],x2[4:]**2, 'r--')

plt.title('Meu Primeiro Grafico!! (É UMA PIRAMIDE)', fontdict={'fontname': 'Ubuntu', 'fontsize': 20})
plt.xlabel('Montanha russa')
plt.ylabel('reto toda vida')

plt.xticks([0,10,20,])
plt.yticks([0,10,20,])


plt.legend()

plt.show
```

SEGUNDO GRAFICO

```python
labels= ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels,values)

bars[0].set_color('g')
patterns = ['.','/','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
#bars[0].set_hatch('.')

plt.figure(figsize=(6,4))

plt.show()
```
